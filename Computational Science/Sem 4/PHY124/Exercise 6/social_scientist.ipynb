```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/merterol/Desktop/iMac27_github/uzh/Computational Science/Sem 4/PHY124/Exercise 6/API_EG.ELC.ACCS.ZS_DS2_en_csv_v2_13544.csv", skiprows=4)
df.head()
```

```python
df = df.dropna(axis='columns', how='all') # Drop columns with all NaN values
df.fillna(method='ffill', inplace=True) # Fill forward to fill NaN values
df.drop(columns=["Indicator Name", "Indicator Code"], inplace=True, errors='ignore') # Drop unnecessary columns

non_countries = ['ARB', 'CEB', 'CSS', 'EAP', 'EAR', 'EAS', 'ECA',
    'ECS', 'EMU', 'EUU', 'FCS', 'HIC', 'HPC', 'IBD', 'IBT', 'IDA',
    'IDB', 'IDX', 'LAC', 'LCN', 'LDC', 'LIC', 'LMC', 'LMY', 'LTE',
    'MEA', 'MIC', 'MNA', 'NAC', 'OED', 'OSS', 'PRE', 'PSS', 'PST',
    'SAS', 'SSA', 'SSF', 'SST', 'TEA', 'TEC', 'TLA', 'TMN', 'TSA',
    'TSS', 'UMC', 'WLD']

df = df[~df['Country Code'].isin(non_countries)] # get rid of country codes that are not countries

df.head()
```

Get one row per (country, year).

```python
df_melted = df.melt(
    id_vars=["Country Name", "Country Code"],  # these stay as identifying columns
    var_name="Year",
    value_name="ElectricityAccess"            # a more descriptive name for our values
)

# Convert 'Year' to integer if possible:
df_melted['Year'] = pd.to_numeric(df_melted['Year'], errors='coerce')
```

Create animated plot

```python
import plotly.express as px

fig = px.choropleth(
    df_melted,
    locations="Country Code",        # uses the ISO 3-letter country codes
    color="ElectricityAccess",       # the data to color-code
    hover_name="Country Name",       # column to show in the tooltip
    animation_frame="Year",          # make an animation by year
    color_continuous_scale=px.colors.sequential.Plasma,  # or any other scale
    range_color=[0, 100],            # because it's a percentage from 0 to 100
    labels={"ElectricityAccess": "Access (%)"}
)

fig.update_layout(
    title="Access to Electricity (% of population) - World Bank",
)
fig.show()
```
