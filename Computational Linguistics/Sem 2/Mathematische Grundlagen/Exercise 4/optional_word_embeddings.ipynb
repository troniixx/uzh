```python
# Load embeddings for the adjectives

import gzip
import csv

# Path to the compressed file
filename = 'embeddings.csv.tar.gz'

# Dictionary to store embeddings
embeddings = {}

# Open the gzipped file and read its contents
with gzip.open(filename, 'rt') as file:  # Open the file in read-text mode
    csv_reader = csv.reader(file)  # Use csv.reader to read the file
    for row in csv_reader:
        row = row[0].split()
        word = row[0]  # The first column contains the word
        if word not in adjectives:
          continue
        vector = [float(value) for value in row[1:]]  # Convert the remaining columns to floats
        embeddings[word] = vector  # Store the vector in the dictionary

print("Loaded embeddings for {} words.".format(len(embeddings)))
```

```python
query = "calm"

# DONE Write code that outputs the adjective with the highest cosine similarity
# to the query (excluding the query itself).
from sklearn.metrics.pairwise import cosine_similarity

similarities = {}

for word in embeddings:
    if word == query:
        continue
    similarity = cosine_similarity([embeddings[query]], [embeddings[word]])[0][0]
    similarities[word] = similarity
    
most_similar = max(similarities, key=similarities.get)

print("The most similar word to '{}' is '{}' with a cosine similarity of {:.2f}.".format(query, most_similar, similarities[most_similar]))
```
