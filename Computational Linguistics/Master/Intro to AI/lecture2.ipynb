# Agent Programs – Task 2

Create and run your own example of a utility-based agent implemented in Python for the following agent
and task environment:
- utility-based vacuum cleaner agent that:
- operates in a 2-location environment (locations 'A' and ‘B’), where each location can be 'Clean' or 'Dirty’ with the probability of being 'Dirty’ is 40% for ‘A’ and 20% for 'B'
- can Suck (clean current location), Move Left, or Move Right
- has a utility function which maximizes cleanliness while minimizing unnecessary movements

```python
import random

P_DIRTY = {"A": 0.4, "B": 0.20}

REWARD_CLEAN = 5
COST_MOVE = 1
COST_SUCK = 0

def others(loc):
    if loc == "A":
        return "B"
    else:
        return "A"

def utility_agent(loc, status):
    if status == "Dirty":
        return "Suck"
    
    p = P_DIRTY[others(loc)]
    expected_gain_move = p * REWARD_CLEAN - COST_MOVE
    
    if expected_gain_move > 0:
        return "Move right" if loc == "A" else "Move left"
    else:
        return "Suck"
    
def run_episode(steps = 6):
    state = {
        "A": "Dirty" if random.random() < P_DIRTY["A"] else "Clean",
        "B": "Dirty" if random.random() < P_DIRTY["B"] else "Clean",
    }
    
    loc = random.choice(["A", "B"])
    utility = 0
    
    for _ in range(steps):
        action = utility_agent(loc, state[loc])
        
        if action == "Suck":
            if state[loc] == "Dirty":
                utility += REWARD_CLEAN
                state[loc] = "Clean"
            utility += COST_SUCK
        elif action == "Move left":
            utility -= COST_MOVE
            loc = "A"
        elif action == "Move right":
            utility -= COST_MOVE
            loc = "B"
            
    return utility, state
```

```python
for i in range(5):
    utility, final_state = run_episode()
    print(f"Episode {i+1}: Utility = {utility}, Final State = {final_state}")
```
